# Airbnb Data Analysis

__In what neighborhood are most Airbnbs located?__

```python
airbnbs["neighbourhood"].value_counts().head(5)
```

__What is the most expensive Airbnb?__

```python
#convert prices (strings) to integers
intPrices = []
for price in airbnbs["price"]:
    if isinstance(price, str):
        intPrices.append(Decimal(sub(r'[^\d.]', '', price)))
airbnbs["price"] = intPrices
```

```python
airbnbs[airbnbs['price'] == max(airbnbs['price'])]['price'].drop_duplicates()
```

__In what neighborhoods are Airbnbs the most expensive (within a $200 range of the max)?__

```python
airbnbs[airbnbs['price'] == range(max(airbnbs['price']))]['neighbourhood'].value_counts()
```

__What is the minimum price for an Airbnb?__

```python
airbnbs[airbnbs['price'] == min(airbnbs['price'])]['price'].drop_duplicates()
```

__In what neighborhoods are Airbnbs the least expensive?__

```python
airbnbs[airbnbs['price'] == min(airbnbs['price'])]['neighbourhood'].value_counts()
```

__Are there any Airbnbs <$20?__

```python
airbnbs[airbnbs["price"] < 20]
```
